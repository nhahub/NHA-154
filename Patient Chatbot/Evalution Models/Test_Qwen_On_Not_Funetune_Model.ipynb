# **Antcoloy Algo**

```python
class LLMEvaluator:
    def __init__(self, pipe, dataset, prompt_template):
        self.pipe = pipe
        self.dataset = dataset
        self.prompt_template = prompt_template

        print("Loading Metrics (BERTScore & ROUGE)...")
        self.bertscore = evaluate.load("bertscore")
        self.rouge = evaluate.load("rouge")

    def evaluate_batch(self, params, indices_list):
        predictions = []
        references = []

        temp = max(0.01, params['temperature'])
        top_p = params['top_p']
        max_tokens = int(params['max_tokens'])

        for idx in indices_list:
            instruction = self.dataset[idx]['instruction']
            reference = self.dataset[idx]['output']
            prompt = self.prompt_template.format(instruction, "")

            try:
                outputs = self.pipe(
                    prompt,
                    max_new_tokens=max_tokens,
                    temperature=temp,
                    top_p=top_p,
                    do_sample=True
                )
                full_text = outputs[0]['generated_text']

                # Clean text based on your specific template markers
                gen_text = full_text.split("<|im_start|>assistant\n")[-1].strip()
                gen_text = gen_text.replace("<|eot_id|>", "").replace("<|im_end|>", "").strip()

                predictions.append(gen_text)
                references.append(reference)

            except Exception as e:
                print(f"Error at index {idx}: {e}")
                predictions.append("")
                references.append(reference)

        if not predictions:
            return {
                "score_used_for_opt": 0.0,
                "bert_mean": 0.0,
                "bert_median": 0.0,
                "bert_max": 0.0,
                "rouge_l": 0.0
            }

        bert_results = self.bertscore.compute(
            predictions=predictions,
            references=references,
            lang="ar"
        )
        f1_scores = bert_results["f1"]

        bert_mean = np.mean(f1_scores)
        bert_median = np.median(f1_scores)
        bert_max = np.max(f1_scores)

        rouge_results = self.rouge.compute(
            predictions=predictions,
            references=references
        )
        rouge_l = rouge_results['rougeL']


        return {
            "score_used_for_opt": bert_median,
            "bert_mean": bert_mean,
            "bert_median": bert_median,
            "bert_max": bert_max,
            "rouge_l": rouge_l
        }
```

```python
class AntColonyOptimizer:
    def __init__(self, evaluator, search_space, config):
        self.evaluator = evaluator
        self.search_space = search_space

        self.num_ants = config.get('num_ants', 5)
        self.num_iterations = config.get('iterations', 3)
        self.evaporation_rate = config.get('evaporation_rate', 0.5)
        self.alpha = config.get('alpha', 2.0)

        self.num_eval_samples = config.get('num_eval_samples', 3)

        total_data = len(evaluator.dataset)
        safe_sample_size = min(total_data, self.num_eval_samples)
        self.test_indices = random.sample(range(total_data), safe_sample_size)

        print(f"Optimization will run on {safe_sample_size} random examples: {self.test_indices}")

        self.pheromones = {}
        for param_name, values in search_space.items():
            self.pheromones[param_name] = {v: 1.0 for v in values}

        self.best_global_score = -1
        self.best_params = {}
        self.best_metrics_display = {}

    def _select_value(self, param_name):
        possible_values = list(self.pheromones[param_name].keys())
        pheromone_values = list(self.pheromones[param_name].values())
        weights = [p ** self.alpha for p in pheromone_values]
        return random.choices(possible_values, weights=weights, k=1)[0]

    def _update_pheromones(self, iteration_results):
        # Evaporate
        for param in self.pheromones:
            for value in self.pheromones[param]:
                self.pheromones[param][value] *= (1 - self.evaporation_rate)

        for result in iteration_results:
            score = result['score']
            for key, value in result['params'].items():
                self.pheromones[key][value] += score

    def display_pheromone_status(self, iteration):
        print(f"\n --- Pheromones (Iter {iteration}) ---")
        for param_name in self.pheromones:
            df = pd.DataFrame(list(self.pheromones[param_name].items()), columns=['Val', 'Pheromone'])
            df = df.sort_values(by='Pheromone', ascending=False).head(3)
            print(f"🔹 {param_name}: {df['Val'].tolist()} (Top preferred)")
        print("-" * 40)

    def optimize(self):
        print("Starting Optimization (Target: Median BERTScore)...")

        for i in range(1, self.num_iterations + 1):
            self.display_pheromone_status(i)
            iteration_results = []

            for ant_id in range(self.num_ants):
                current_params = {k: self._select_value(k) for k in self.search_space}

                metrics = self.evaluator.evaluate_batch(current_params, self.test_indices)

                opt_score = metrics['score_used_for_opt']
                mean_val = metrics['bert_mean']
                max_val = metrics['bert_max']
                rouge_val = metrics['rouge_l']

                iteration_results.append({'params': current_params, 'score': opt_score})

                print(f"   🐜 Ant {ant_id+1}: {current_params}")
                print(f"      ➤ Median BERT: {opt_score:.4f} (Used for Best)")
                print(f"      ➤ Mean BERT:   {mean_val:.4f} | Max BERT: {max_val:.4f}")
                print(f"      ➤ ROUGE-L:     {rouge_val:.4f}")

                # Update Global Best based on MEDIAN
                if opt_score > self.best_global_score:
                    self.best_global_score = opt_score
                    self.best_params = current_params
                    self.best_metrics_display = metrics
                    print(f"      🌟 NEW BEST FOUND (By Median)!")

            self._update_pheromones(iteration_results)

        print("\n Optimization Finished!")
        print(f" Best Parameters: {self.best_params}")
        print(f" Final Best Metrics:")
        print(f"   Median BERT: {self.best_global_score:.4f}")
        print(f"   Mean BERT:   {self.best_metrics_display.get('bert_mean', 0):.4f}")
        print(f"   Max BERT:    {self.best_metrics_display.get('bert_max', 0):.4f}")
        print(f"   ROUGE-L:     {self.best_metrics_display.get('rouge_l', 0):.4f}")

        return self.best_params
```

```python
import random
random.seed(52)
search_space = {
    'temperature': [0.1,0.2,0.4, 0.3, 0.5, 0.7, 0.9, 1.0],
    'top_p':       [0.5, 0.7, 0.9, 0.95, 0.99],
    'max_tokens':  [128, 256, 512, 1024]
}

aco_config = {
    'num_ants': 5,
    'iterations': 4,
    'num_eval_samples': 20,
    'evaporation_rate': 0.3,
    'alpha': 2.0
}

evaluator = LLMEvaluator(pipe, test_dataset, prompt_template)
optimizer = AntColonyOptimizer(evaluator, search_space, aco_config)

best_params = optimizer.optimize()
```
